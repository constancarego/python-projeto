```python
# Importando bibliotecas necessárias

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data do projeto obtida atraves da plataforma PRODES (http://terrabrasilis.dpi.inpe.br/app/dashboard/deforestation/biomes/legal_amazon/rates)
# Data utilizada tem o nome de "Taxas de desmatamento acumulado por ano- Amazonia legal- Estados" e ficheiro CVS está disponivel no reportorio phyton-projeto na conta de GitHub @constancarego assim como este codigo

# Lendo o arquivo CVS e carregando os dados em um DataFrame

df=pd.read_csv("C:\\Users\\const\\Desktop\\python\\terrabrasilis_legal_amazon_20_8_2023_1695218846036.csv", sep=';')

# Renomeando colunas para facilitar o entendimento

df = df.rename(columns={'area km²': 'area', 'uf': 'district'})

# Convertendo a 'area' para valores inteiros e ajustando a escala para 100

for i in range(len(df['area'])):    
    df['area'][i] = int(df['area'][i].replace('.', '').replace(',', ''))/100

# Calculando a area total queimada por distrito

total_area_by_district = df.groupby('district')['area'].sum().reset_index()

# Plotando um grafico de barras para mostrar a area total queimada por distrito

total_area_by_district.plot(x="district", y="area", kind="bar",title='Total Area Burnt', figsize=(12,6))

# Calculando a area total queimada por ano

total_area_by_year = df.groupby('year')['area'].sum().reset_index()

# Plotando um grafico de linhas para mostrar a area total queimada ao longo dos anos

total_area_by_year.plot(x="year", y="area", kind="line",title='Total Area Burnt by District', figsize=(12,6))

# Configurando parametros para melhorar a apresentaçao dos graficos

plt.figure(figsize=(10, 6)) 

# Criando um boxplot para visualizar a distribuiçao da area queimada por distrito

sns.boxplot(data=df, x='district', y='area')
plt.title('Box Plot of Burnt Area by District')
plt.xlabel('District')
plt.ylabel('Burnt Area in Km2')
plt.xticks(rotation=0)

# Ajustando o layout para melhorar a estica dos graficos

plt.tight_layout()
```
